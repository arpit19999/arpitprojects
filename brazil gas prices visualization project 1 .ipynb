```python
df.rename(
    columns={
        "DATA INICIAL": "start_date",
        "DATA FINAL": "end_date",
        "REGIÃO": "region",
        "ESTADO": "state",
        "PRODUTO": "product",
        "NÚMERO DE POSTOS PESQUISADOS": "no_gas_stations",
        "UNIDADE DE MEDIDA": "unit",
        "PREÇO MÉDIO REVENDA": "avg_price",
        "DESVIO PADRÃO REVENDA": "sd_price",
        "PREÇO MÍNIMO REVENDA": "min_price",
        "PREÇO MÁXIMO REVENDA": "max_price",
        "MARGEM MÉDIA REVENDA": "avg_price_margin",
        "ANO": "year",
        "MÊS": "month",
        "COEF DE VARIAÇÃO DISTRIBUIÇÃO": "coef_dist",
        "PREÇO MÁXIMO DISTRIBUIÇÃO": "dist_max_price",
        "PREÇO MÍNIMO DISTRIBUIÇÃO": "dist_min_price",
        "DESVIO PADRÃO DISTRIBUIÇÃO": "dist_sd_price",
        "PREÇO MÉDIO DISTRIBUIÇÃO": "dist_avg_price",
        "COEF DE VARIAÇÃO REVENDA": "coef_price"
    },
    inplace=True
)
df.dtypes
```

```python
# convert the data columns
df['dist_avg_price']=pd.to_numeric(df['dist_avg_price'],errors='coerce')
df['dist_sd_price']=pd.to_numeric(df['dist_sd_price'],errors='coerce')
df['dist_min_price']=pd.to_numeric(df['dist_min_price'],errors='coerce')
df['dist_max_price']=pd.to_numeric(df['dist_max_price'],errors='coerce')
df['coef_dist']=pd.to_numeric(df['coef_dist'],errors='coerce')
```

```python
df.dtypes
```

```python
df.query('year! =2019 & product in ["GLV","GNV"]').groupby(['year','region'])['avg_price'].agg('sum')
```

```python
df.head()
```

```python
# plot Regionwise Yearwise Average Price data
fig,ax=plt.subplots(figsize=[14,6])
df.query('year!=2019 & product in ["GLV","GNV"]' ).groupby(['year','region'])['avg_price'].sum().unstack().plot(ax=ax)
plt.grid(True)
```

```python
# plot Regionwise Yearwise Average Price Changes (% wise) data 
fig,ax=plt.subplots(figsize=[15,5])
df.query('year!=2019 & product in ["GLV","GNV"]').groupby(['year','region']).sum()['avg_price'].pct_change().unstack().plot(ax=ax)
plt.grid(True)
```

```python
df.query('year!=2019 & region in ["NORDESTE"]').groupby(['year','state'])['avg_price'].agg('sum')
```

```python
# plot Regionwise Yearwise Average Price Changes (% wise) data 
fig,ax=plt.subplots(figsize=[15,6])
df.query('year!=2019 & region in ["NORDESTE"]').groupby(['year','state'])['avg_price'].agg('sum').pct_change().unstack().plot(ax=ax)
plt.grid(True)
```

```python
fig,ax=plt.subplots(figsize=[15,6])
df.query('year!=2019 & region in ["CENTRO OESTE"]').groupby(['year','state'])['avg_price'].agg('sum').unstack().pct_change().plot(ax=ax)
plt.grid(True)
```

```python
fig,ax=plt.subplots(figsize=[15,7])
df.query('year!=2019 ').groupby(['year','region']).sum()['avg_price'].pct_change().unstack().plot(ax=ax)
plt.grid(True)
```
